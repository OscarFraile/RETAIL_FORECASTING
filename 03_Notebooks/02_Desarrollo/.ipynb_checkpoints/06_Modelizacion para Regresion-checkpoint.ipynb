## MODELIZANDO PARA UN PRODUCTO

```python
#Cogemos un solo producto de una tienda para hacer la modelización y captar el proceso para replicarlo en el resto
#Pasamos de un x con 21800 registros a un x con 836 registros
x = x.loc[(x.store_id == 'CA_3') & (x.item_id == 'FOODS_3_586')]
```

```python
x
```

```python
y = y[y.index.isin(x.index)]
```

### MODELIZAR

#### Definir las variables de modelización

```python
var_modelizar = x.columns.to_list()[3:]
```

#### Definir la validacion

```python
#CUIDADO! Debemos hacer la validación cruzada para series temporales 
    #Hay que pasarle TimeSeriesSplit(3, --> Número de veces que repetiremos el proceso de test de los datos
                        #test_size = 8) --> Tamaño de grupo de test. 8 días en este caso
time_cv = TimeSeriesSplit(3, test_size = 8)
```

#### Crear el pipe y el diccionario de algorimos, parámetros y valores a testar

Modificar para dejar solo los algoritmos que se quieran testar.

Modificar los parámetros.

```python
#Isaac lo deja con los datos por defecto porque ya ha probado diferentes combianciones y ha visto que la mejor es esta, pero 
#en un caso real tendríamos que ir probando las combinaciones y tomar la decisión
#Podríamos hacerlo con varios algoritmos pero sabemos que en este tipo de proyectos es este algoritmo el que funciona bien
pipe = Pipeline([('algoritmo',HistGradientBoostingRegressor())])

grid = [ {'algoritmo': [HistGradientBoostingRegressor()],
#          'algoritmo__learning_rate': [0.01,0.025,0.05,0.1],
#          'algoritmo__max_iter': [50,100,200],
#          'algoritmo__max_depth': [5,10,20,50],
#          'algoritmo__scoring': ['neg_mean_absolute_error'],
#          'algoritmo__l2_regularization': [0,0.25,0.5,0.75,1]
         }
       ]
```

#### Optimizar los hiper parámetros

#####  Con random search

```python
#Podemos cambiar el numero de iteraciones y la métrica para que nos de el scoring por si consiguieramos mejores resultados
random_search = RandomizedSearchCV(estimator = pipe,
                                   param_distributions = grid, 
                                   n_iter = 3, 
                                   cv = time_cv, #Esta validación cruzada es la específica de series temporales
                                   scoring = 'neg_mean_absolute_error', 
                                   verbose = 0,
                                   n_jobs = -1)

modelo = random_search.fit(x[var_modelizar],y)

pd.DataFrame(random_search.cv_results_).sort_values(by = 'rank_test_score')
#Debemos mirar el valor de la columna mean_test_score para compararlo con AUC que nos salga sobre val_y
```

```python
#Entrenamos el modelo con los parámetros óptimos sobre todos los datos para conseguir el mejor modelo
modelo_final = modelo.best_estimator_.fit(x[var_modelizar],y)
modelo_final
```

```python
#Hacemos una predicción sobre los mismos datos solamente para saber si el proceso es correcto aunque la predicción en si no
#nos servirá porque estamos prediciendo sobre los mismos datos que hemos entrenado y saldrá posiblemente sobre ajustada 
prediccion = modelo_final.predict(x[var_modelizar])
```

```python
#Generamos un df temporal para visualizar los resultados
temp = pd.concat([x.date.reset_index(drop=True),y.reset_index(drop=True),pd.Series(prediccion,name = 'prediccion')],axis = 1).set_index('date')
```

## ESCALAR A TODOS LOS PRODUCTOS

### MODELIZACIÓN

Ya tenemos el proceso funcionando para modelizar un producto.

Ahora tenemos que escalarlo a todas las combinaciones de tiendas y productos.

Primero vamos a volver a importar los datos.

```python
nombre_x = 'x_preseleccionado.pickle'
nombre_y = 'y_preseleccionado.pickle'
```

```python
x = pd.read_pickle(ruta_proyecto + '/02_Datos/03_Trabajo/' + nombre_x)
y = pd.read_pickle(ruta_proyecto + '/02_Datos/03_Trabajo/' + nombre_y)
```

Incluímos como una nueva variable la combinación de tienda-producto y eliminamos las originales porque queremos predecir los resultados de los 10 productos en 2 tiendas = 20 productos

```python
#Esto lo incluiremos en las funciones del código de preproducción
x.insert(loc=0,column='producto',value=x.store_id + '_'+ x.item_id)
x = x.drop(columns = ['store_id','item_id'])
```

Para poder aplicar el proceso de forma recurrente vamos a crear una lista donde cada elemento sean los datos de cada producto.

```python
productos = list(x.groupby("producto"))
productos
```

Encapsulamos el proceso de modelización individual en una función, a la que iremos llamando en cada producto.

```python
productos[0][1]
```

```python
#Recibe los datos de las x y la y de cada producto

def modelizar(x_producto, y):
    #x_producto es el resultado de la combinación de los productos con cada tienda      
    #Excluye el producto como variable de modelización. Cuenta desde la posición 2 xq la posición 1 es la variable producto 
    var_modelizar = x_producto.columns.to_list()[2:]
    
    #Define la validación cruzada
    time_cv = TimeSeriesSplit(3, test_size = 8)
    
    #Define la parrilla de algoritmos
    pipe = Pipeline([('algoritmo',HistGradientBoostingRegressor())])
    grid = [ 
         {'algoritmo': [HistGradientBoostingRegressor()]
#          'algoritmo__learning_rate': [0.01,0.025,0.05,0.1],
#          'algoritmo__max_iter': [50,100,200],
#          'algoritmo__max_depth': [5,10,20,50],
#          'algoritmo__scoring': ['neg_mean_absolute_error'],
#          'algoritmo__l2_regularization': [0,0.25,0.5,0.75,1]
         }
                       
    ]
           
    #Crea los modelos
    random_search = RandomizedSearchCV(estimator = pipe,
                                   param_distributions = grid, 
                                   n_iter = 1, 
                                   cv = time_cv, 
                                   scoring = 'neg_mean_absolute_error', 
                                   verbose = 0,
                                   n_jobs = -1)
    
    modelo = random_search.fit(x_producto[var_modelizar],y)
    
    #Reentrena el mejor sobre todos los datos
    modelo_final = modelo.best_estimator_.fit(x_producto[var_modelizar],y)
    
    #Devuelve como salida el modelo final
    return(modelo_final)
```

Creamos otra lista vacía donde vamos a ir guardando los modelos entrenados de cada producto.

```python
lista_modelos = []
```

Recorremos cada elemento de la lista de productos y vamos rellendando lista_modelos con el nombre y modelo final de cada producto.

```python
y
```

```python
#Productos: Es una lista que tiene una tupla de dos elementos y contiene los nombres de cada producto y sus datos
#    1º es el nombre del producto 
#    2º son los datos asociados a ese nombre


for cada in productos:
    #La función dice: Por cada uno de los elementos que hay en productos, coge el primer elemento y lo llamas 'producto' 
    #y al segundo lo llamas 'datos'
    producto = cada[0]
    datos = cada[1]
    
    #Selecciona solo los datos de la target que se corresponden con el producto porque hay mas registros de taget que de productos
    y_cada = y[y.index.isin(datos.index)]

    #Llama a la funcion de modelizar. Ahora x=datos, y=y_cada
    modelo = modelizar(datos,y_cada)
    #Añade el modelo final a la lista
    lista_modelos.append((producto,modelo))

#Este bucle repítelo hasta que termine toda la lista de productos
```

```python
lista_modelos
```

Comprobamos.

NOTA: El objetivo aquí no es evaluar la calidad del modelo, ya que no estamos usando los datos de entrenamiento.

Es simplemente comprobar que el proceso funciona bien y no vemos nada raro antes de continuar.

```python
#unificar el df para poder extraer los x e y de cada producto
df =  pd.concat([x,y],axis=1)

for cada in range(0,len(lista_modelos)):
    
    #Prepara los datos necesarios
    producto = lista_modelos[cada][0]
    modelo = lista_modelos[cada][1]
    #Nosotros haremos una preselección de variables específicada para cada modelo y es la combinación de variables con la
    #que entrenaremos el modelo y que después deberemos pasarle para la predicción al modelo específico para el que está 
    #pensado porque en caso contrario dará un error. Esta línea lo que hace es extraer el nombre del producto y el modelo 
    #con la combinación de variables 
    variables = modelo[0].feature_names_in_
    target = 'ventas'

    #Definimos x, y
    x_prod = df.loc[df.producto == producto].drop(columns=target).copy()
    y_prod = df.loc[df.producto == producto,'ventas'].copy()

    #Incluimos una columna con la fecha
    date = x_prod.date.values
    
    #Genera graficos de predicho vs real de cada modelo
    predicciones = pd.DataFrame(data={'date': date,
                                      'producto': producto,
                                      'ventas': y_prod,
                                      'prediccion': modelo.predict(x_prod[variables])})

    predicciones['prediccion'] = predicciones.prediccion.astype('int')
    predicciones.set_index('date',inplace=True)

    predicciones.tail(100).plot(figsize = (16,8),title = producto);
    
#Este bucle repítelo hasta que termine toda la lista de lista_modelos 
```
